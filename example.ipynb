```python
from sklearn.datasets import load_breast_cancer
import pandas as pd
import matplotlib.pyplot as plt
from boxplots import boxplots
```

```python
data = load_breast_cancer()
feature_matrix = pd.DataFrame(data.data, columns=data.feature_names)
target = pd.DataFrame(data.target)

target[target == 0] = data.target_names[0]
target[target == 1] = data.target_names[1]
```

```python
plt.rcParams["figure.figsize"] = [16, 9]
boxplots(feature_matrix, target)
plt.show()
```

```python
plt.rcParams["figure.figsize"] = [12, 15]
boxplots(
    feature_matrix,
    target,
    order=["malignant", "benign"],
    colors={"benign": "g", "malignant": "r"},
    ncol_subplots=7,
)
plt.show()
```
